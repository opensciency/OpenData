You can add content to the module using Jupyter notebooks, if you wish. This allows you to display code cells with their output, or take advantage of the rich display options that Jupyter provides.

Suppose you want to talk about the increase in use of preprints as a form of rapid communication and Green Open Access publication. You want to show a graph, and transparently show how the graph is produced from an open data set.

In this demo, we import `pandas`, and we load a dataset of preprint usage per month, from {cite:t}`polka_jessica_k_2020_3955154`, then we keep only some relevant columns and change the column labels so they are shorter when displayed in the legend.

```python
import pandas as pd
```

The data is available publicly on Google Sheets. Here we load the data into a `pandas` dataframe using the correct sheet name from the multi-sheet source, then we display the column labels.

```python
sheet_id = "1c1WzYIoLhzY09AkwyhSVDtan3oYV8qrHa3WgPHAeVVM"
sheet_name = "Preprints%20per%20month"
url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}"
```

```python
preprint_data = pd.read_csv(url)
```

```python
preprint_data.columns
```

Notice that we set the cell above to hide the output, which is a bit long. You can hide input or output of code cells at will to make the displayed content more readable. We won't go into details here, but let the team know if you'd like help with this.

We will plot the data from the first few columns, and rename the columns with a shorter label, for display in the plot legend, and finally display a stacked filled-line plot with the [`.plot.area()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.area.html) method for dataframes.

```python
preprint_data = preprint_data.filter(['Source',
       'arXiv q-bio Through 2018-12, PrePubMed; from 2019-01, arxiv.org',
       'F1000 & Open Research platforms Sum of columns T+U',
       'PeerJ Preprints Through 2018-12, PrePubMed; From 2019-01, europePMC (includes revisions)',
       'bioRxiv Through 2018-12, PrePubMed; from 2019-01, EuropePMC (includes revisions)',
       'preprints.org EuropePMC', 'ChemRxiv EuropePMC', 'Sneak Peek Website',
       'Research Square Crossref (filtered for v1 preprints)- 2020-04 from EuropePMC',
       'The Lancet Website', 'medRxiv EuropePMC'], axis=1)
```

```python
preprint_data.columns = ['Source',
       'arXiv q-bio',
       'F1000',
       'PeerJ Preprints',
       'bioRxiv',
       'preprints.org EuropePMC', 'ChemRxiv EuropePMC', 'Sneak Peek Website',
       'Research Square',
       'The Lancet Website', 'medRxiv EuropePMC']
```

```python
ax = preprint_data.plot.area(figsize=(10,6));
```

And that is one interesting plot showing an explosion in preprint use in the last few years!
